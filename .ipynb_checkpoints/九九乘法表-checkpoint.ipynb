```python
#直行---方法1 
for i in range(1, 10):
    print('\t'.join([('%s*%s=%s') % (j,i,i*j) for j in range(1,10)]))
    
```

```python
#直行---方法2
print('\n'.join(['\t'.join(['%d*%d=%d' % (j,i,i*j) for j in range(1,10)]) for i in range(1,10)]))
```

```python
#橫列---方法1
for j in range(1, 10):
    print('\t'.join([('%s*%s=%s') % (j,i,i*j) for i in range(1,10)]))
    
```

```python
#橫列---方法2
print('\n'.join(['\t'.join(['%d*%d=%d' % (j,i,i*j) for i in range(1,10)]) for j in range(1,10)]))
```

```python
#全部在一行
for i in range(1, 10, 1):
    for j in range(1, 10, 1):
        print(i, 'x', j, '=', i*j)
```
